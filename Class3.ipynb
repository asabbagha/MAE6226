```python
import math
import numpy
from matplotlib import pyplot
```

```python
N = 50 #number of points
x_start, x_end = -2.0,2.0 #x boundaries
y_start, y_end = -1.0,1.0 #y boundaries
x = numpy.linspace(x_start,x_end, N)
y = numpy.linspace(y_start, y_end, N)

print('x= ',x)
print('y= ',y)

X, Y = numpy.meshgrid(x,y) #creates mesh grid combined of x and y
```

```python
%matplotlib inline

width = 10.0
height = (y_end - y_start) / (x_end - x_start) * width
pyplot.figure(figsize=(width, height)) #sets size of figure
pyplot.xlabel('x,' )
pyplot.ylabel('y,' )
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.scatter(X, Y, s=5, marker='o');
```

```python
strength_source = 5.0
x_source, y_source = -1.0,0.0 #location of source

#compute velocity field of mesh grid
u_source = (strength_source / (2*math.pi) * 
           (X-x_source) / ((X-x_source)**2 + (Y - y_source)**2))
v_source = (strength_source / (2*math.pi) * 
           (Y - y_source) / ((X - x_source)**2 + (Y - y_source)**2))
```

```python
#plotting stream lines
width = 10.0
height = (y_end - y_start) / (x_end - x_start)*width
pyplot.figure(figsize = (width,height))
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_source, v_source,
                 density = 2, linewidth = 1, arrowsize = 2, arrowstyle = '->')
pyplot.scatter(x_source, y_source,
              s = 80, marker='o');
```

```python
strength_sink = -20.0
x_sink, y_sink = 2.0,0.0

#compute vel of sink
u_sink = (strength_sink / (2*math.pi) * 
         (X - x_sink) / ((X-x_sink)**2 + (Y - y_sink)**2))
v_sink = (strength_sink / (2*math.pi) *
         (Y - y_sink) / ((Y - y_sink)**2 + (X - x_sink)**2))
```

```python
#plot sink streamlines
width = 10.0
height = (y_end - y_start) / (x_end - x_start) * width
pyplot.figure(figsize=(width,height))
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.streamplot(X, Y, u_sink, v_sink,
                 density = 2, linewidth = 1, arrowsize = 2, arrowstyle ='->')
pyplot.scatter(x_sink, y_sink, s=80, marker='o');
```

```python
#combine source and sink
u_comb = u_source + u_sink
v_comb = v_source + v_sink

width = 10
height = (y_end - y_start) / (x_end - x_start) * width
pyplot.figure(figsize=(width,height))
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.streamplot(X, Y, u_comb, v_comb,
                 density = 2, linewidth = 1, arrowsize = 2, arrowstyle='->')
pyplot.scatter([x_source, x_sink] , [y_source, y_sink],
              s=80, marker='o');
```
